```python
import re


def clean_text(text):

    # 1. Fix words broken across lines
    text = re.sub(
        r"(\w)-\n(\w)",
        r"\1\2",
        text
    )

    # 2. Detect likely paragraph boundaries
    text = re.sub(
        r"([.!?])\n(?=[A-Z])",
        r"\1\n\n",
        text
    )

    # 3. Convert remaining single newlines into spaces
    text = re.sub(
        r"(?<!\n)\n(?!\n)",
        " ",
        text
    )

    # 4. Normalize spaces
    text = re.sub(
        r"[ \t]+",
        " ",
        text
    )

    return text.strip()


def remove_repeated_header(text):

    text = re.sub(
        r"^\d+\s+Christoph Mattmann\s+et al\.\s+/"
        r"\s+Procedia Computer Science 270 \(2025\)"
        r"\s+1517–1526\s*",
        "",
        text
    )

    return text.strip()
```

```python
pages = []

for page_number, page in enumerate(doc):

    raw_text = page.get_text("text")

    cleaned_text = clean_text(raw_text)

    cleaned_text = remove_repeated_header(cleaned_text)

    pages.append({
        "page_number": page_number + 1,
        "text": cleaned_text
    })
```

```python
all_paragraphs = []

for page in pages:

    paragraphs = page["text"].split("\n\n")

    for paragraph in paragraphs:

        paragraph = paragraph.strip()

        if paragraph:

            all_paragraphs.append({
                "page_number": page["page_number"],
                "text": paragraph
            })
```

```python
MAX_CHUNK_SIZE = 2000

chunks = []

current_chunk = []
current_size = 0


def save_current_chunk():

    if current_chunk:

        chunks.append({
            "text": "\n\n".join(
                item["text"]
                for item in current_chunk
            ),

            "pages": sorted(
                set(
                    item["page_number"]
                    for item in current_chunk
                )
            )
        })


for paragraph in all_paragraphs:

    text = paragraph["text"]
    text_length = len(text)

    # Very large paragraph
    if text_length > MAX_CHUNK_SIZE:

        save_current_chunk()

        chunks.append({
            "text": text,
            "pages": [paragraph["page_number"]]
        })

        current_chunk = []
        current_size = 0

    # Paragraph fits
    elif current_size + text_length <= MAX_CHUNK_SIZE:

        current_chunk.append(paragraph)
        current_size += text_length

    # Paragraph does not fit
    else:

        save_current_chunk()

        # Overlap only if the current chunk
        # contains multiple paragraphs
        if len(current_chunk) > 1:

            last_paragraph = current_chunk[-1]

            current_chunk = [last_paragraph]

            current_size = len(
                last_paragraph["text"]
            )

        else:

            current_chunk = []
            current_size = 0


save_current_chunk()
```

```python
import numpy as np

query = "What is CapsNet?"

query_embedding = model.encode(query)

def cosine_similarity(a, b):

    return np.dot(a, b) / (
        np.linalg.norm(a) *
        np.linalg.norm(b)
    )
```

```python
# 1. User's question
query = "What is CapsNet?"

# 2. Convert the question into an embedding
query_embedding = model.encode(query)

# 3. Compare query with every chunk
results = []

for chunk in chunks:

    score = cosine_similarity(
        query_embedding,
        chunk["embedding"]
    )

    results.append({
        "chunk_id": chunk["chunk_id"],
        "text": chunk["text"],
        "pages": chunk["pages"],
        "score": score
    })

# 4. Sort from most relevant to least relevant
results = sorted(
    results,
    key=lambda x: x["score"],
    reverse=True
)

# 5. Retrieve the top 3 chunks
top_results = results[:3]

# 6. Combine retrieved chunks into context
context = "\n\n".join(
    result["text"]
    for result in top_results
)

# 7. Build the augmented prompt
prompt = f"""
Use the following context to answer the question.

Context:
{context}

Question:
{query}

Answer only using the provided context.
"""

# 8. Inspect the prompt
print(prompt)
```
